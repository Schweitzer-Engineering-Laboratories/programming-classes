```python
def sub(x):
    print("Your submarine has surfaced in a deep sea cavern. "
         "There is only a faint illumination from stones imbeded in the walls of the cave.\n")
    options = ["Look around", "Call out", "Return"]
    action = input("You hear a resounding screech beneath the surface of the water. What do you do? "
                   + str(options))
    while action not in options:
        print("You can't just sit around, try again: " + str(options)
             )
        action = input()
    if action==options[0]:
        print("\nSomething slithers through the water. As it sufaces, you see its toothed spine and hear a constant hiss.")
        x = 1
    if action==options[1]:
        print("\nYour voice resounds throughout the room. A clicking noise follows...")
        x = 2
    if action==options[2]:
        print("\nYou return to the submarine. As you try to resubmerge, something wraps around the vessel, preventing movement.\n"
             "From the look out, you can see a long, blackened creature coiling its body around the metal. It pulls tighter digging in.\n"
             "You see sparks before the submarine is crushed with you inside.\n\nYou died.\n")
    return x

def noise(x):
    options = ["Apprehension", "Curiousity", "Excitement"]
    action = input("How do you react to the noise? "
                   + str(options))
    
    while action not in options:
        print("You must be feeling something, try again: " + str(options))
        action = input()
    if action==options[0]:
        print("\nThe monster senses your fright. It approaches and reveals rows of teeth. You try and flee but the slick surface of the cave causes you to slip.\n"
             "\nYou die from the shock of being skewered. The leviathan has eaten you.\n")
        x = 1
    if action==options[1]:
        print("\nAs you touch the waters surface, you feel something slimy running under your fingers. It slowly creeps out of the water. The soft body tighens as it reaches your forearm--yanking hard.\n"
             "\nYou are drowned, pulled into the cold abyss.\n")
        x = 2
    if action==options[2]:
        print("\nThe beast emerges, its dark body wetting the cavern floor. You nearly slip into the seemingly bottomless depths as you lose your footing.\n"
             "\nIt catches you in a soft carress. The touch loving but cold from the frigid water. It coos something you can't understand, echoing in the faint light.\n")
        x = 3
    return x

def rom(x):
    options = ["Caress it", "Push it", "Kiss it"]
    action = input("Show the sea creature how you feel: "
                   + str(options))
    x = 0
    while action not in options:
        print("The leviathan is anticipating your response... " + str(options))
        action = input()
    if action==options[0]:
        print("\nAs you touch its body, covering your hand in slime, you can feel its eyeless gaze on you. It nesstles your forhead. You feel that your affection has come across to the leviathan.\n"
             "\nAlthough you aren't sure if it knows your language, it can understand the sincerity. It wraps a tentacle higher around your body, the sharp comb tickling you as it screeches. No doubt, if you could understand the leviathan, you would know it was saying 'I love you.'\n")
        x = 2
    if action==options[1]:
        print("\nTrying to free yourself from its grasp, the leviathan growls. The sound stands your hair in fright, its looming body forcing the air out of your lungs as it tightens its hold on you. "
             "Before asphyxiate, one of the serrated tenticles nicks your body as it coils around you. You feel it digging in until it reaches your neck, cutting your jugular.\n\nYou have bled out.\n")
        x = 1
    if action==options[2]:
        print("\nThe leviathan shys away. Its body almost returns to the water but something draws it back to you. It wasn't expecting you to be so bold.\n"
             "\nAs it makes it back, a tendril reaches towards you. You take its hand and clicks echo from it, overjoyed to have your love.\n")
        x = 2
    return x

def leviathan():
    x = 0
    r = 0
    i = 0

    e = sub(x)

    if (e == 1) or (e == 2):
        e = noise(r)
        if e == 3:
            r = rom(i)
            if r == 1:
                e = 1
            else:
                e = 3

    else:
        e = 1

    print("\nYears later... ")

    if e == 3:
        print("you have lost contact with the world, never to return home. "
             "In the depths of the sea, you have made a new home with your loving spouse, your other half, the leviathan. "
             "It treats you well and provides what you need to the best of its abilities.\n"
             "\nGood end.")
    if e == 1:
        print("you body has gone undiscovered. Multiple search parties have been sent but none have returned with anything conclusive, if at all.\n\nBad end.")
    if e == 2:
        print("the bones of your body, picked clean by the sea life, are discovered by an expedition.\n\nBad end.")
```

```python
from IPython.display import Image
from IPython.core.display import HTML
from IPython.core.display import display

def Speed():
    k=0
    while k==0:
        k=1
        action=input("How fast do you want to go? (in knots)")
        try:
            is_it_a_number=int(action)
        except ValueError:
            print("\n",
                  "Please enter a proper number.")
            k=0
    while int(action)>30:
        print("\n", 
              "Your submarine cannot go faster than 30 knots!")
        k=0
        while k==0:
            k=1
            action=input()
            try:
                is_it_a_number=int(action)
            except ValueError:
                print("Please enter a proper number.")
                k=0
    if int(action)<=30 and int(action)>=20:
        print("\n","You find atlantis in no time!", flush=True)
        display(Image(url="https://i.pinimg.com/originals/41/18/04/411804d9e7c6eb101798da6d5ac7da92.jpg"))
    if int(action)<20:
        print("\n","You travel too slow and you end up getting caught in a volcanic eruption. "
              "\n","Your gold melts and turns you into a trophy!")

def Atlantis():
    options = ["explore", "nothing"]
    action = input("What do you do next?"+str(options))
    if action == "explore":
        print("\n","You get back in your submarine and you enter in the coordinates to Atlantis")
        Speed()
    elif action == "nothing":
        print("\n","I have had enough adventure for one day!")
        print("\n","You are locked out of your submarine.......dun dun DuuuUN!")
    else:
        print("\n","Dunnuh, dunnuh, dunnuh, dunnuh, dunnuh, dunnuh, dunnuh,.......dunnuh dunnuh DuuuUN!")
        print("\n","A shark comes and bites off your arm with all of the treasure!")

def password():
    options = ["sun"]
    action = input("Please enter the password to access my database:")
    while action not in options:
        print("\n","Nope! Try again. HINT: I like this mass of plasma")
        action = input()
    if action=="sun":
        print("\n","Access Granted")

def Gold():
    options = ["explore", "turn back"]
    action = input("You have found a shipwreck with a creepy, creepy face. What would you like to do?"+str(options))
    if action == "explore":
        print("\n","You discover gold!! Among the pieces of gold, we find a map to the long lost city of Atlantis!")
        Atlantis()
    elif action == "turn back":
        print("\n","An eel attacks you and shocks your ship, frying all of your controls. You drift aimlessly until you run out of air.")
    else:
        print("\n","An eel attacks you and shocks your ship, frying all of your controls. You drift aimlessly until you run out of air. Think quicker next time.")

def TheNextThing():
    options = ["Go to the surface", "Find the Merchant Royal", "Let's find the bottom of this here pond!", 
               "Head back to the dock", "Search for mermaids"]
    action = input("Where do you want to go?"+"\n"+str(options))
    gas=2
    while action not in options:
        print("\n",".......I am not sure where that is....Try somewhere else.")
        action = input()
    while gas>0:
        if action=="Go to the surface":
            action = input("The weather looks nice! Now where would you like to go now? "+"\n"+str(options[1:5]))
            gas=gas-1
        elif action == "Find the Merchant Royal":
            Gold()
            gas=-1
        elif action == "Let's find the bottom of this here pond!":
            print("\n","The submarine implodes....WAFOOM!")
            gas=-1
        elif action == "Head back to the dock":
            options.remove("Head back to the dock")
            action = input("The dock is being under attack by pirates! We need to act fast, where would you like to go? "+"\n"+str(options))
            gas=gas-1
        else: 
            action == "Search for mermaids"
            print("\n","You loose track of time and even though you hear singing your sub is never found. dun dun daaaaaa")
            gas=-1
    if gas==0:
        print("\n","Run out of gas")

def sunken_ship():
    print("\n","You are on a deep-ocean dive, the surface team controlling your sub goes silent.", flush=True)
    print("\n","That's okay, your sub has a text interface.", flush=True)
    options = ["yes", "no"]
    action = input("Would you like to communicate with the sub by using the computer? " + str(options))
    while action not in options:
        print("\n","Error, error! You have entered an invalid command. Please try again.")
        action = input()
    if action=="yes":
        print("\n","Hello, my name is Olaf the submarine, and I like the sun! ")
        password()
        TheNextThing()
    else:
        print("\n","You slowly run out of air. The lights go off. You are all A L O N E!")
```
