```python
import time
def desert():
    print("After a long trek through the desert", flush=True)
    options=["rest","keep going"]
    action = input("light is starting to fade and your tired and weary"
                    + str(options))
    while action not in options:
        print("Invalid Option")
        action = input()
    if action=="keep going":
        print("you pass out due to exhaustion")
    else:
        print("you make camp and sleep")
    time.sleep(2)    
    print("you wake up to find several buzzards flying above you")
    battle()
def battle():
    time.sleep(2)
    print("You see a rock and a bone")
    time.sleep(2)
    options = ["rock", "bone", "hands"]
    action = input("What do you choose...(rock, bone or hands)\n")
    while action not in options:
        print("You entered an invalid option, try again")
        action = input()
    if action=="rock":
        print("You throw it and misses.")
        time.sleep(2)
        print("It claws your arms")
        time.sleep(2)
        outcome = "You die of blood loss"
        time.sleep(2)
        print(outcome)
    elif action=="bone":
        outcome = "You hit and kill it"
        time.sleep(2)
        print(outcome)
    elif action=="hands":
        print("It claws your arms")
        time.sleep(2)
        outcome = "You die of blood loss"
        print(outcome)
    if outcome=="You die of blood loss":
        time.sleep(2)
        print("GAME OVER")
    elif outcome == "You hit and kill it":#options from battle
        time.sleep(2)
        print("CONGRATULATIONS, you've slain the buzzard.\n")
        time.sleep(2)
        escape()
def ending():
    options3 = ["investigate", "keep going"]
    action3 = input("Would you like to keep going or investigate?"
                            + str(options3))
    while action3 not in options3:
        print("You entered an invalid option, try again")
        action3 = input()
    if action3=="keep going":
        print("You arrive to your destination, Your body is pleading for water, but you made it. \nWas it worth it? \nDid that buzzard deserve to die? \nYeah, absolutely. It attacked you. \Great job!")
    else:
        print("Your body pleading with your mind, you give in. \nThat water looks so good. \nHas water ever looked so good? \nYou get closer...and closer... and closer before-\nYou're trapped? In quicksand? \nNo, no, no, it can't be. \nYou hear a cawing in the distance. \nYou look up. \nPanic sets in and the cawing gets closer, and louder, until...\nThe end.")
def escape():
    options = ["search for water", "camp", "continue on"]
    action = input("You're exhausted, out of breathe, and your body is craving water. \nWhat do you do?"
                    + str(options))
    while action not in options:
        print("You entered an invalid option, try again")
        action = input()
    if action=="search for water":
        print("You can survive three days without water. \nBut can you survive a night in the desert? \nThe answer's no. \nYou wander through the night on a search for water, but your confidence was far too high. \nThe sunrises and you realize that you're lost and still have no water in sight. \nYou die of dehydration.")
    elif action=="camp":
        print("You've decided to camp for the night. Good choice. \nYou wake up thirsty, but you're alive.")
        options2 = ["yes", "no"]
        action2 = input("Would you like to continue?"
                        + str(options2))
        while action2 not in options2:
            print("You entered an invalid option, try again")     
        if action2=="yes":
            print("You continue on to your final destination.\nAs the sun rises, you see the sun shimmer off of something in the distance. \nIs it water? \nHave your prayers been answered?")
           # ending()
        else:
            print("Uh, okay. You stand completely still like a lunatic until you die. The end?")
    else:
        print("You've decided to hike through the night. \nAs the sunrises, you see the sun shimmer off of something in the distance. \nIs it water? \nHave your prayers been answered?")
    ending()
```
