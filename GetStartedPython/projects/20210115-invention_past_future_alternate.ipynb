### Invention

```python
def invention():
    input("\"Time Machine\" \n PRESS ENTER OR ENTER A PROMPT TO CONTINUE")
    input("You are standing in front of an impressive chrome plated machine.")
    input("You are nearly finished with the invention that you know will change the world")
    options = ["Mcguffin", "Flux Capacitor", "Fork","Translucent Floating Sphere"]
    for x in range(4):
        action = input("You wisper to yourself.\"Just needs a few more components\" \n PLEASE CHOOSE AN ITEM."
                     + str(options))
        while action not in options:
            print("Hmm, that didn't seem to work. Try again")
            action = input()
        if action=="Mcguffin":
                print("Some lights turned on, \"I think something's happening.\"")
                options.remove("Mcguffin")
        elif action=="Flux Capacitor":
                print("Some lights turned on, \"I think something's happening.\"")
                options.remove("Flux Capacitor")
        elif action=="Translucent Floating Sphere":
                input("The Sphere didn't seem to do anything. \"How strange.\"")
                options.remove("Translucent Floating Sphere")
        elif action == "Fork":
            input("That was left over from dinner! \"Now, why would I do that? \"")
    if options==["Fork"]:
        input("Start the Machine?")
        input("The machine starts right up.")
        input("You enter the machine")
        input("lights flash, and you feel a sense of vertigo")
        input("Are you ready?")
```

```python
def fight():
    options = ["Kill Them With Kindness", "Blow Their Robots Minds With Dad Jokes"]
    action = input("How would you like to Fight"+ str(options))
    if action not in options:
        print("recalculating...try again") 
        
    elif action =="Kill Them With Kindness":
        print("You walk up to the biggest Robot, flash your megawatt smile and say Does someone need a hug?")
                
    elif action =="Blow Their Robots Minds With Dad Jokes":
        print ("your time machine gets stuck in reverse and releases in the middle of a global pandemic") 
         
    else:
        print("You wake up and realize you are back in 2021 and have just received the COVID shot")


def join():
    action=input("You decide to join the robot army. You're taken back to their compound. What do you say? (Take me to your leader, Assimilate me)")
    print(action)
    if action=="Take me to your leader":
        actiona=input("You demand to be taken to the robot leader in the toughest voice you can manage. \n Hundreds of blank, lifeless eyes just stare at you.\n What do you do? (panic, explore)")
        if actiona=="panic":
            print("You panic, and run right into a trap door, falling down into a damp pit. You die.")
        else:
            actionb=input("You explore the compound and discover a well-hidden doorway. What do you do, (go back, try door)")
            if actionb=="go back":
                print("You go back to the entrance and realize something is happening, all of the doors have vanished. You die")
            else:
                print("The door opens on its own, the lock was on the inside. There is an egg man, the most horrific of beasts. You get yoked to death.")
    elif action=="Assimilate me":
        print("You gave up easily. You die for no reason")
    else:
        print(" I don't know that command, you die.")
        

def hide():
    num_robots = int(input("A robot army?! How many are there?"))
    if num_robots >= 50:
        print("I count", num_robots, "robots, there are too many for me to take down by myself! \nI am going back to the time machine and getting out of here!")
    elif num_robots < 50:
        print("I count", num_robots, ", but they are all focused on the twinkie factory.\nI think I will hide until they leave and then I can take what's left.\nMaybe they will also have HoHos.")
    else:
        print("I only count", num_robots, ".\nI am going to hide behind this burger joint and get something to eat.")
        print("The food was made of nuts and bots. You can't eat it and die.")
        
def future():
    choice = input("Welcome to the 3028, you have encountered a robot army! \n Do you want to hide, join them, or fight them?")
    if choice == "fight them":
         fight()
    elif choice == "join them":
        join()
    elif choice == "hide":
        hide()
    else:
        print("You didn't make it to the time machine or entered the wrong answer and the robots obliterated you!")
```
