```python
def kmd():
    forest = 1
    print("You enter a foresty forest")

    while forest < 8:

    #first input loop    
        while forest == 1:
            if forest == 1:
                print("------------------------------------")
                print("Wolves start to chase you!")
                second = input("What do you do?: ")
                print("------------------------------------")
            if second.lower() == ("flee"):
                forest = 2
            elif second.lower() == ("fight"):
                print("------------------------------------")
                print("You quickly perish!")
                print("-----------------------------------")
            elif second.lower() == ("look around"):
                forest = 3
            else:
                print("-----------------------------------")

        #second input loop    
        while forest == 2:
            if forest == 2:
                print("------------------------------------")
                print("You run several steps and find bread")
                bread = input("Do you eat the bread?: ")
                print("------------------------------------")
            
            if bread.lower() == ("yes"):
                print("------------------------------------")
                print("Wolves eat you while you eat bread!")
                print("-----------------------------------")
                forest = 1
            if bread.lower() == ("no"):
                forest = 1
            else:
                forest = 2

        #third input loop
        while forest == 3:
            if forest == 3:
                print("You find a sword in a stone!")
                sword = input("Do you take the sword?: ")
            
            if sword.lower() == ("yes"):
                print("------------------------------------")
                print("You defeat the alpha male and claim the pack as your own.")
                forest = 4
            elif sword.lower() == ("no"):
                print("------------------------------------")
                print("You quickly perish!")
                print("------------------------------------")
                forest = 1

        #fourth input loop
        while forest == 4:
            if forest == 4:
                print("You now have a wolf pack with very cute puppies in it and reached a checkpoint!")
                print("---------------------------------------------")
                pack = input("what do you do? HINT: you can pet the puppies for hunt with them: ")
            
            if pack.lower() == ("hunt"):
                print("------------------------------------")
                print("You find a small set of track and bigger set of tracks.")
                forest = 5
            if pack.lower() == ("keep going"):
                print("------------------------------------")
                print("You quickly perish!")
                print("------------------------------------")
                forest = 3
            if pack.lower() == ("pet puppies"):
                print("------------------------------------")
                print("You quickly perish!")
                print("------------------------------------")
                forest = 3
            else:
                print("------------------------------------")

        #fifth input loop
        while forest == 5:
            if forest == 5:
                track = input("which set of tracks do you follow?: ") 
                print("------------------------------------")
            if track.lower() == ("smaller"):
                print("------------------------------------")
                print("You follow the smaller set of tracks")
                print("------------------------------------")
                forest = 6
            elif track.lower() == ("bigger"):
                print("------------------------------------")
                print("You follow the bigger set of tracks")
                print("------------------------------------")
                forest = 7

        #sixth input loop
        while forest == 6:
            if forest == 6:
                print("You encounter a cougar with her cubs and instantly get mawled but you defended your wolf pack and they got away.")
                print("------------------------------------")
                print("GG")
                forest = 9

        #seventh input loop
        while forest == 7:
            if forest == 7:
                print("You encounter a giant grizzly bear, catching him off guard. He gets angry and mawls you while your wolf pack runs away.")
                print("------------------------------------")
                print("GG")
                forest = 9

        if forest == 9:
            print("Goodbye")
```
